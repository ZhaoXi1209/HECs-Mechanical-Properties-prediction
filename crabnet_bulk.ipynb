```python
data_train = rf'data/materials_data/elastic_modulus/bulk/train.csv'
```

```python
cb_bulk.load_data(data_train, batch_size=2**9, train=False)
```

```python
output_train_bulk = cb_bulk.predict(cb_bulk.data_loader)
```

```python
df_bulk_train = pd.DataFrame(
    {'calc': output_train_bulk[0], 'pred': output_train_bulk[1], 'formula': output_train_bulk[2],'uncert':output_train_bulk[3]})
```

```python
df_bulk_train.to_csv("data/materials_data/elastic_modulus/bulk/df_bulk_train.csv")
```

```python
sns.set(style="ticks")
plt.figure(figsize=(8,8),dpi=100)  ### 修改分辨率

axs = sns.scatterplot(data=df_bulk_train, x= "calc", y="pred",s=120,marker='o',alpha = 0.8, color = 'b')
axs.tick_params(labelsize = 18)
formatXy = 'y = x'
axs.plot((0, 1), (0, 1), transform=axs.transAxes, linewidth=2, ls='--',c='k',alpha=0.4, label = formatXy)

act = np.array(output_train_bulk[0])
pred = np.array(output_train_bulk[1])
    
r2 = r2_score(act, pred)
mae = mean_absolute_error(act, pred)
rmse = np.sqrt(mean_squared_error(act, pred))

str = 'R2 = %.3f \nRMSE = %.3f \nMAE = %.3f '%(r2, rmse, mae)

axs.set_xlim(200,400)
axs.set_ylim(200,400)

axs.legend(loc = 'lower right', fontsize = 18)
axs.text(210,360,str,fontsize = 16)

plt.tick_params(left='on',bottom='on',direction='in')
plt.xlabel("DFT Calculation modulus(GPa)", fontdict={'size': 18})
plt.ylabel("Predict modulus(GPa)", fontdict={'size': 18})

plt.savefig('/mnt/e/ZHAO/Data/ELASTIC_HEC/Figure/bulk_crabnet.png', dpi = 600, bbox_inches = 'tight')   ### 保存图片
```
